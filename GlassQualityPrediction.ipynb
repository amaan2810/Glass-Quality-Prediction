```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('D:/IMS Proschool/Business Analytics/Practice/Self/Glass_Quality_Participants_Data/Train.csv')
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data['class'] = data['class'].replace(2,0)
data.head()
```

```python
data.grade_A_Component_1.unique()
```

```python
data.grade_A_Component_2.unique()
```

```python
data = data.drop(['grade_A_Component_2'], axis=1)
```

```python
features = data.drop(['class'], axis=1)
label = data['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(features, label, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X_train = std.fit_transform(X_train)
X_val = std.transform(X_val)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.svm import SVC
```

```python
lr = LogisticRegression()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier(n_estimators=100, max_features=5)
knn = KNeighborsClassifier(n_neighbors=3)
nb = GaussianNB()
xgb = XGBClassifier()
svm = SVC(kernel='linear', probability=True, C=7)
```

```python
lr_score = cross_val_score(lr, X_train, y_train, cv=5, scoring='neg_log_loss')
dt_score = cross_val_score(dt, X_train, y_train, cv=5, scoring='neg_log_loss')
rf_score = cross_val_score(rf, X_train, y_train, cv=5, scoring='neg_log_loss')
knn_score = cross_val_score(knn, X_train, y_train, cv=5, scoring='neg_log_loss')
nb_score = cross_val_score(nb, X_train, y_train, cv=5, scoring='neg_log_loss')
xgb_score = cross_val_score(xgb, X_train, y_train, cv=5, scoring='neg_log_loss')
svm_score = cross_val_score(svm, X_train, y_train, cv=5, scoring='neg_log_loss')
```

```python
print('lr',lr_score.mean())
print('dt',dt_score.mean())
print('rf',rf_score.mean())
print('knn',knn_score.mean())
print('nb',nb_score.mean())
print('xgb',lr_score.mean())
print('svm',lr_score.mean())
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import ExtraTreesClassifier

gbc = GradientBoostingClassifier(n_estimators=100, max_features=8)
etc = ExtraTreesClassifier(n_estimators=80, max_features=7)

gbc_score = cross_val_score(gbc, X_train, y_train, cv=5, scoring='neg_log_loss')
etc_score = cross_val_score(etc, X_train, y_train, cv=5, scoring='neg_log_loss')

print('gbc',gbc_score.mean())
print('etc',etc_score.mean())
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import StratifiedKFold

bc = BaggingClassifier(base_estimator=ExtraTreesClassifier(n_estimators=80, max_features=7), random_state=0)

scores = cross_val_score(bc, X_train, y_train, scoring = 'neg_log_loss', cv = StratifiedKFold(n_splits = 5, shuffle = True))

print(scores.mean())
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import StratifiedKFold

bc = BaggingClassifier(base_estimator=ExtraTreesClassifier(n_estimators=80, max_features=7), random_state=0)

scores = cross_val_score(bc, features, label, scoring = 'neg_log_loss', cv = StratifiedKFold(n_splits = 5, shuffle = True))

print(scores.mean())
```
